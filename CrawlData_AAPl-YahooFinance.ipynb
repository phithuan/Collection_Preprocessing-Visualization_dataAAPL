```python
import yfinance as yf
import pandas as pd
```

```python
# Ticker của Apple trên Yahoo Finance
aapl = yf.Ticker("AAPL")
```

```python
# 1. Lấy thông tin cơ bản về công ty
info = aapl.info
print("Thông tin cơ bản của Apple:")
for key, value in info.items():
    print(f"{key}: {value}")
```

```python
# Lấy bảng dòng tiền (Cash Flow Statement)
cash_flow = aapl.cashflow
print("\nBảng dòng tiền (Cash Flow Statement) của Apple:")
print(cash_flow)

# Lưu bảng dòng tiền vào file CSV
cash_flow.to_csv("apple_cash_flow.csv")
```
