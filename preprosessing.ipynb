```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("AAPL.xlsx", sheet_name= 'AAPL - Cash Flow FY')
data.head()
```

```python
data.isnull().values.sum()
```

```python
# xóa các hàng chứa null
df = data.dropna()
df.head()
```

```python
df.isnull().values.sum()
```

```python
row_as_column = df.iloc[0]  # Lấy hàng có chỉ số 2 (hàng thứ 3)
row_as_column
```

```python
# Đặt lại tên các cột

# Đổi tên các cột trong DataFrame
df.rename(columns=row_as_column, inplace=True)
# Xóa hàng với chỉ số là 3
df.drop(index=3, inplace=True)
# Hiển thị DataFrame 
df.head(4)
```

```python
df.info()
```

```python
# Lưu DataFrame vào file CSV nếu cần
df.to_csv('CashFlow.csv', index=False)
```

```python
"""list(df.axes) trong Pandas được dùng để lấy danh sách các trục (axes) của một DataFrame. Mỗi DataFrame có hai trục:

Trục hàng (Index): Đây là danh sách các hàng (chỉ số) trong DataFrame.
Trục cột (Columns): Đây là danh sách các tên cột trong DataFrame."""
list(df.axes)
```
